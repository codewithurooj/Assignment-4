```python
import random

def computer_guess(max_val):
    low = 1
    high = max_val
    feedback = ''


    while feedback != 'c':

        if low != high:
            guess = random.randint(low, high)
        else:
            guess = low


        feedback = input(f"Is {guess} too high (H), too low (L), or correct (C)? ").lower()

        if feedback == 'h':
            high = guess - 1
        elif feedback == 'l':
            low = guess + 1

    print(f"Yay! The computer guessed your number: {guess}")

def main():
    print("Think of a number between 1 and 100 and I'll guess it!")
    input("Press Enter when you're ready...")
    computer_guess(100)

if __name__ == "__main__":
    main()
```
