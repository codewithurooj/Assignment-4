```python
def main():
    print("Welcome to the Mad Libs Story Generator!")

    # Gather words from the user.
    adjective1 = input("Enter an adjective: ")
    noun1 = input("Enter a noun: ")
    verb1 = input("Enter a verb: ")
    adjective2 = input("Enter another adjective: ")
    noun2 = input("Enter a plural noun: ")
    verb_ing = input("Enter a verb ending in -ing: ")
    adverb = input("Enter an adverb: ")
    place = input("Enter a place: ")

    # Build the Mad Libs story using string formatting.
    story = (
        f"Once upon a time in {place}, there was a {adjective1} {noun1} who loved to {verb1}.\n"
        f"Every day, it would watch {adjective2} {noun2} {verb_ing} {adverb} in the streets.\n"
        "It was a sight that made everyone smile and inspired countless adventures!"
    )

    # Display the completed story.
    print("\nHere's your Mad Libs story:")
    print(story)


if __name__ == "__main__":
    main()










```
