```python
import random

def get_random_word():
    words = ["python", "hangman", "project", "challenge", "coding"]
    return random.choice(words).upper()

def play_hangman():
    word = get_random_word()
    # Create a list with the same length as the word, initially filled with underscores.
    guessed_word = ["_"] * len(word)
    guessed_letters = set()
    tries = 6

    print("Welcome to Hangman!")
    print(" ".join(guessed_word))
    print(f"You have {tries} tries.")

    while tries > 0 and "_" in guessed_word:
        guess = input("Enter a letter: ").strip().upper()

        # Check if letter has already been guessed.
        if guess in guessed_letters:
            print(f"You already guessed '{guess}'.")
            continue

        guessed_letters.add(guess)

        # If the guess is in the word, reveal its positions.
        if guess in word:
            for index, letter in enumerate(word):
                if letter == guess:
                    guessed_word[index] = guess
            print("Correct guess!")
        else:
            tries -= 1
            print(f"Wrong guess! Tries remaining: {tries}")

        # Display the current progress (guessed word).
        print(" ".join(guessed_word))

    if "_" not in guessed_word:
        print(f"Congratulations! You guessed the word: {word}")
    else:
        print(f"Game over! The word was: {word}")

if __name__ == "__main__":
    play_hangman()
```
