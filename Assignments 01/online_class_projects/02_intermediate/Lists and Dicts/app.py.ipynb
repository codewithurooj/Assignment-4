```python
def main():
    # -------------------------
    # Problem #1: List Practice
    # -------------------------
    print("Problem #1: List Practice")
    # Create a list called fruit_list that contains the following fruits.
    fruit_list = ['apple', 'banana', 'orange', 'grape', 'pineapple']
    # Print the length of the list.
    print("Initial length of the fruit list:", len(fruit_list))
    # Add 'mango' at the end of the list.
    fruit_list.append('mango')
    # Print the updated list.
    print("Updated fruit list:", fruit_list)

    # -------------------------
    # Problem #2: Index Game
    # -------------------------
    print("\nProblem #2: Index Game")
    # Initialize a sample list with mixed elements.
    sample_list = [42, "hello", 3.14, "world", True]
    print("Initial list:", sample_list)

    # Game interaction loop
    while True:
        # Present the user with the available operations.
        print("\nSelect an operation:")
        print("1. Access an element")
        print("2. Modify an element")
        print("3. Slice the list")
        print("4. Exit")
        choice = input("Enter the number of your choice: ").strip()

        if choice == "1":
            # --- Accessing an Element ---
            try:
                index = int(input("Enter the index to access: "))
            except ValueError:
                print("Invalid input. Please enter an integer for the index.")
                continue
            result = access_element(sample_list, index)
            print("Result:", result)

        elif choice == "2":
            # --- Modifying an Element ---
            try:
                index = int(input("Enter the index to modify: "))
            except ValueError:
                print("Invalid input. Please enter an integer for the index.")
                continue
            new_value = input("Enter the new value: ")
            result = modify_element(sample_list, index, new_value)
            if isinstance(result, list):
                print("Updated list:", result)
            else:
                print(result)

        elif choice == "3":
            # --- Slicing the List ---
            try:
                start_index = int(input("Enter the start index: "))
                end_index = int(input("Enter the end index (not included): "))
            except ValueError:
                print("Invalid input. Please enter integers for the indices.")
                continue
            result = slice_list(sample_list, start_index, end_index)
            if isinstance(result, list):
                print("Sliced list:", result)
            else:
                print(result)

        elif choice == "4":
            # Exit the game
            print("Goodbye!")
            break

        else:
            # Invalid operation selection.
            print("Invalid option. Please enter 1, 2, 3, or 4.")


def access_element(lst, index):
    """
    Accepts a list and an index.
    Returns the element at the specified index or an error message if out-of-range.
    """
    if 0 <= index < len(lst):
        return lst[index]
    else:
        return "Index out of range."


def modify_element(lst, index, new_value):
    """
    Accepts a list, an index, and a new value.
    Replaces the element at the specified index with the new value.
    Returns the updated list or an error message if the index is out-of-range.
    """
    if 0 <= index < len(lst):
        lst[index] = new_value
        return lst
    else:
        return "Index out of range."


def slice_list(lst, start, end):
    """
    Accepts a list, a start index, and an end index.
    Returns a new list with elements from start up to (but not including) end.
    If indices are out-of-range (or start > end), returns an error message.
    """
    if start < 0 or end > len(lst) or start > end:
        return "Slice indices are out of range."
    return lst[start:end]


if __name__ == "__main__":
    main()
```
