```python
import random

def main():
    # Number of rounds to play.
    NUM_ROUNDS = 5
    score = 0

    # Welcome the player.
    print("Welcome to the High-Low Game!")
    print("--------------------------------")

    # Play the game for the designated number of rounds.
    for round_number in range(1, NUM_ROUNDS + 1):
        print(f"\nRound {round_number}")

        # Milestone #1: Generate random numbers for you and the computer.
        player_number = random.randint(1, 100)
        computer_number = random.randint(1, 100)

        # Show the player's number.
        print(f"Your number is {player_number}")

        # Milestone #2: Get the user's guess.
        guess = input("Do you think your number is higher or lower than the computer's?: ").strip().lower()
        # Safeguard user input for valid answers.
        while guess not in ["higher", "lower"]:
            guess = input("Please enter either higher or lower: ").strip().lower()

        # Milestone #3: Check the guess.
        # If the player and computer have the same number, it's considered incorrect.
        if guess == "higher":
            if player_number > computer_number:
                print("You were right!")
                score += 1
            else:
                print("Aww, that's incorrect.")
        else:  # guess == "lower"
            if player_number < computer_number:
                print("You were right!")
                score += 1
            else:
                print("Aww, that's incorrect.")

        # Show the computer's number and the current score.
        print(f"The computer's number was {computer_number}")
        print(f"Your score is now {score}")

    # End of game message.
    print("\nThanks for playing!")
    # Conditional ending messages based on performance.
    if score == NUM_ROUNDS:
        print("Wow! You played perfectly!")
    elif score >= NUM_ROUNDS // 2:
        print("Good job, you played really well!")
    else:
        print("Better luck next time!")

if __name__ == "__main__":
    main()
```
