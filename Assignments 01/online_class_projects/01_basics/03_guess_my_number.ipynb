```python
import random

def main():
    # Generate the secret number at random (from 0 to 99 inclusive)
    secret_number = random.randint(0, 99)
    print("I am thinking of a number between 0 and 99...")

    # Get the initial guess from the user
    guess = int(input("Enter a guess: "))

    # Keep prompting the user until they guess correctly
    while guess != secret_number:
        if guess < secret_number:
            print("Your guess is too low")
        else:
            print("Your guess is too high")

        # Prompt for a new guess on a new line
        guess = int(input("\nEnter a new number: "))

    # Congratulate the user once the correct number is guessed
    print("Congrats! The number was: " + str(secret_number))

if __name__ == '__main__':
    main()
```
