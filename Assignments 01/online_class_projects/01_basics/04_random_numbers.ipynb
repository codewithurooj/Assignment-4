```python
import random

N_NUMBERS = 10
MIN_VALUE = 1
MAX_VALUE = 100

def main():
    for _ in range(N_NUMBERS):
        # Generate a random number in the specified range and print it followed by a space.
        print(random.randint(MIN_VALUE, MAX_VALUE), end=" ")
    # Print a newline after all numbers are printed.
    print()

if __name__ == '__main__':
    main()
```
