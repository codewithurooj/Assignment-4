```python
def main():
    # Use a for loop to count down from 10 to 1.
    for i in range(10):
        # The current number is 10 minus the number of times the loop has run.
        print(10 - i, end=" ")
    # After the loop, print "Liftoff!".
    print("Liftoff!")

if __name__ == "__main__":
    main()
```
