```python
MINIMUM_HEIGHT = 50  # Set the minimum height requirement

def main():
    while True:
        height = input("How tall are you? ")  # Prompt the user for height
        if not height:  # Stop the program if no input is provided
            break
        height = float(height)  # Convert height to a number
        if height >= MINIMUM_HEIGHT:
            print("You're tall enough to ride!")
        else:
            print("You're not tall enough to ride, but maybe next year!")

if __name__ == "__main__":
    main()
```
