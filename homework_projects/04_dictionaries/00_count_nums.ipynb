```python
def main():
    numbers = []
    while True:
        user_input = input("Enter a number: ")
        if user_input == "":
            break
        numbers.append(int(user_input))
    counts = {}
    for num in numbers:
        if num in counts:
            counts[num] += 1
        else:
            counts[num] = 1
    for num, count in counts.items():
        print(f"{num} appears {count} times.")

if __name__ == "__main__":
    main()
```
