```python
def main():
    # Initialize the list of numbers
    numbers = [1, 2, 3, 4]

    # Loop through each element and double it
    for i in range(len(numbers)):
        numbers[i] *= 2

    # Print the modified list
    print(numbers)

# Ensure the main function runs when the program is executed
if __name__ == "__main__":
    main()
```
