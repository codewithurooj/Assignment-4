```python
MAX_LENGTH = 3

def shorten(lst):
    while len(lst) > MAX_LENGTH:
        last_elem = lst.pop()  # Remove the last element
        print(last_elem)  # Print the removed element

def get_lst():
    lst = []
    elem = input("Please enter an element of the list or press Enter to stop: ")
    while elem != "":
        lst.append(elem)
        elem = input("Please enter an element of the list or press Enter to stop: ")
    return lst

def main():
    lst = get_lst()
    shorten(lst)

if __name__ == "__main__":
    main()

```
