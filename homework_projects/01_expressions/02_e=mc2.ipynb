```python
# Speed of light in meters per second (constant)
C = 299792458

def main():
    # Continually read mass input from the user
    while True:
        try:
            # Ask the user for mass in kilograms
            mass_in_kg = float(input("Enter kilos of mass (or type 'exit' to quit): "))

            # Calculate the energy using the formula E = m * C^2
            energy_in_joules = mass_in_kg * (C ** 2)

            # Display the results
            print("\ne = m * C^2...")
            print(f"m = {mass_in_kg} kg")
            print(f"C = {C} m/s")
            print(f"{energy_in_joules} joules of energy!\n")
        except ValueError:
            print("Invalid input. Please enter a numeric value for mass.")
            break  # Exit the loop if the user types 'exit'


if __name__ == "__main__":
    main()
```
