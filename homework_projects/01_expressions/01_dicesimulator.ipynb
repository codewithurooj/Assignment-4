```python
# Import the random module to simulate dice rolls
import random

# Define the number of sides on a die
NUM_SIDES = 6

# Function to roll two dice and print the results
def roll_dice():
    # Local variables inside the function
    die1:int = random.randint(1, NUM_SIDES)
    die2 :int= random.randint(1, NUM_SIDES)
    print(f"Die 1: {die1}, Die 2: {die2}, Total: {die1 + die2}")

# Main function to control the flow of the program
def main():
    # Local variable in the main() function
    die1 = 10  # This won't affect the roll_dice() function
    print(f"Starting value of die1 in main(): {die1}")

    # Simulate rolling two dice three times
    roll_dice()
    roll_dice()
    roll_dice()

    # Show that the local die1 variable in main() hasn't changed
    print(f"Ending value of die1 in main(): {die1}")

# Ensure the program runs when executed directly
if __name__ == "__main__":
    main()
```
