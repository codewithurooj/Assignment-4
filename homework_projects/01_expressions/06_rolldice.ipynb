```python
# Import the random library to generate random numbers
import random

def main():
    # Simulate rolling the first die
    die1 = random.randint(1, 6)

    # Simulate rolling the second die
    die2 = random.randint(1, 6)

    # Calculate the total of both dice
    total = die1 + die2

    # Print the results
    print(f"First die roll: {die1}")
    print(f"Second die roll: {die2}")
    print(f"Total of both dice: {total}")


if __name__ == "__main__":
    main()
```
