```python
# Constant beginning of the sentence
SENTENCE_START = "Mad Libs are fun! I used Python to make my "

def main():
    # Prompt the user for an adjective, noun, and verb
    adjective = input("Please type an adjective and press enter: ")
    noun = input("Please type a noun and press enter: ")
    verb = input("Please type a verb and press enter: ")

    # Create and print the Mad Libs sentence
    sentence = SENTENCE_START + f"{adjective} {noun} {verb}!"
    print(sentence)

# Call the main function
if __name__ == "__main__":
    main()
```
